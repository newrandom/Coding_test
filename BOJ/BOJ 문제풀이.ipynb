```python
#21921          (시간초과)
# N : 1 일부터 N일까지
# X : 위의 범위에서 X일 동안
qst = input().split()       # N , X의 값을 입력받아 스플릿함.

days = int(qst[0])        # 1일부터 N일 까지
during = int(qst[1])     # X일 동안, 0<=X<=N

user = input().split()  # 방문자 입력값을 스플릿하여 저장

number = []
for n in user:   # 방문자 입력값을 num 리스트변수에 int로 저장하기
    number.append(int(n))   


plus = 0     # 소계, during동안의 값을 합산한다.
maxSum = 0  # 최대값 , during동안의 값들 중 최대값을 저장한다.
cnt = 0

n = 0

while n < days:
    if during <= days:      #  during의 값이 N값보다 작거나 같으면 진행하고, 높으면 종료
        num = number[n:during]
        plus = sum(num)     # [n:during] 의 합계
        # cnt = 1             # 최대값의 개수 1 설정.
        
        if maxSum < plus:   # plus의 값이 maxSum 보다 크다면
            cnt = 1         # cnt를 1로 리셋
            maxSum = plus   # maxSum = plus로 반환
        elif maxSum == plus:    # maxSum 이 plus와 같다면
            cnt += 1        # cnt 에 1을 더함

        during += 1         # during 의 값을 1 추가함 > 한칸씩 뒤로 옮겨지는 것.

    n += 1                  # n의 값에 1을 추가함 > 한칸씩 앞으로 옮겨지는 것.

if maxSum != 0 :
    print('{}'.format(maxSum))
    print('{}'.format(cnt))
else :
    print("SAD")



# 첫날부터 N일까지의 모든 방문자 수에서 X일동안의 최대 방문자의 합을 구하는 문제.
# 방문자 리스트에서 X일동안의 방문자 합계가 루프되어야 한다.
# 최대 방문자 합계가 몇번 나왔는지도 출력되어야 한다.
```

```python
#2581 정답      메모리 30840 // 시간 736
M = int(input())    # 시작값
N = int(input())    # 마지막 값

p_num_list = []      # 소수를 담을 빈 리스트

def p_num(int):      # 소수찾기 함수
    if int == 1:        # 1은 소수가 아님
        return False

    elif int == 2:      # 2는 소수
        return int
    
    for i in range(2, int):     # 주어진 값을 2부터 int 앞까지의 값으로 나눴을 때, 
        if int % i == 0:        # 나누어 떨어지면
            return False        # 소수가 아니다
    
    return int                  # 나누어 떨어지지 않으면 int를 반환한다.

for num in range(M, N+1):       # M 부터 N까지의 범위에서 루프
    if p_num(num) == False:     # 소수찾기 함수의 값이 False이면
        continue                # 다음으로 이어간다.

    else:                       # int를 반환 받으면
        p_num_list.append(p_num(num))   # 소수를 담는 리스트에 해당 값을 넣는다.

if len(p_num_list) == 0:        # 소수 리스트의 길이가 0 (아무것도 없을때)
    print(-1)                   # -1을 출력한다.

else:                           # 소수 리스트의 길이가 0보다 크면 
    print(sum(p_num_list))      # 소수 리스트 값들의 합을 출력하고,
    print(min(p_num_list))      # 소수 리스트의 최소값을 출력한다.
```

```python
import sys
take = sys.stdin.readline()

def get_decimal(num):
    
    if num != 0:
    
        cnt = 0

        for i in range(num, num*2+1):
            if i == 2:
                cnt += 1

            for j in range(2, i):
                if i % j == 0:
                    break

                else :
                    cnt += 1
                    
        
        print(cnt)

get_decimal(3)
                
```
